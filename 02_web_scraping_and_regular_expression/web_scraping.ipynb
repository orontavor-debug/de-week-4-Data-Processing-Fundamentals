```python
from bs4 import BeautifulSoup  # Import the BeautifulSoup class from the bs4 library for parsing HTML

# Define a string containing HTML code for a product card
html_doc = """
<html>
<title>Product</title>
<body>
<div class="p1">
  <h2>Product Name</h2>
  <span class="price">$59.99</span>
</div>
<div class="p2">
  <h2>Item Name</h2>
  <span class="price">$60.99</span>
</div>
</body>
</html>
"""

# Create a BeautifulSoup object to parse the HTML
# 'html.parser' tells BeautifulSoup to use Python's built-in HTML parser
soup = BeautifulSoup(html_doc, 'html.parser')

div_object = soup.find('div', class_ = 'p2')
# Find the first <h2> tag in the HTML (contains the product name)
title_tag = div_object.find('h2')

# Find the first <span> tag with class 'price' (contains the product price)
price_tag = div_object.find('span', class_='price')

# Extract the text inside the <h2> tag if it exists, otherwise return "N/A"
product_name = title_tag.get_text() if title_tag else "N/A"

# Extract the text inside the <span class="price"> tag if it exists, otherwise return "N/A"
product_price = price_tag.get_text() if price_tag else "N/A"

# Print the extracted product name and price in a readable format
print(f"Product Name: {product_name}")
print(f"Product Price: {product_price}")
```

```python
from bs4 import BeautifulSoup  # Import BeautifulSoup for parsing HTML

# Define HTML containing multiple product cards
html_doc = """
<div class="product-card">
  <h2>Product Name</h2>
  <span class="price">$59.99</span>
</div>
<div class="product-card">
  <h2>Item Name</h2>
  <span class="price">$60.99</span>
</div>
"""

# Create a BeautifulSoup object to parse the HTML
soup = BeautifulSoup(html_doc, "html.parser")

# Find all <div> tags with class "product-card"
cards = soup.find_all("div", class_="product-card")

# Loop through each product card
for c in cards:
    # Find the <h2> tag inside the current card, extract its text, and remove extra spaces
    title = c.find("h2").text.strip()
    
    # Find the <span> tag with class "price" inside the current card, extract its text, and remove extra spaces
    price = c.find("span", class_="price").text.strip()
    
    # Print the product name and price in a readable format
    print(f"{title}: {price}")
```

```python
import requests  # Import requests to make HTTP requests to websites
from bs4 import BeautifulSoup  # Import BeautifulSoup to parse HTML

url = "http://quotes.toscrape.com"
r = requests.get(url)  # Send a GET request to the URL and get the HTML response
soup = BeautifulSoup(r.text, "html.parser")  # Parse the HTML content of the page

quotes = soup.find_all("span", class_="text")  # Find all <span> tags with class "text" (the quotes)
authors = soup.find_all("small", class_="author")  # Find all <small> tags with class "author" (the authors)

data = [{"quote": q.text, "author": a.text} for q, a in zip(quotes, authors)]

for row in data[:3]:  # Loop through the first 3 items in the data list
    print(row)  # Print each quote-author pair
```

```python
import requests, time  # Import requests for HTTP requests and time to pause between requests
from bs4 import BeautifulSoup  # Import BeautifulSoup to parse HTML

base = "http://quotes.toscrape.com"  # Base URL of the website
url = base  # Start scraping from the homepage
all_quotes = []  # Initialize an empty list to store all collected quotes
while True:  # Infinite loop to keep scraping pages until there are no more
    soup = BeautifulSoup(requests.get(url).text, "html.parser")  # Fetch current page HTML and parse it

    for q in soup.select(".quote"):  # Loop through all elements with class "quote" (each quote block)
        all_quotes.append({
            "text": q.select_one(".text").text,  # Extract the text of the quote
            "author": q.select_one(".author").text  # Extract the author name
        })

    nxt = soup.select_one("li.next > a")  # Find the "next" button link
    if not nxt: break  # If there is no next page, exit the loop
    url = base + nxt.get("href")  # Update the URL to the next page
    time.sleep(0.5)  # Wait for 0.5 seconds before the next request (politeness)
print("Collected:", len(all_quotes))  # Print the total number of quotes collected
```

open("quotes.csv", "w", newline="", encoding="utf-8") opens a file named quotes.csv for writing with UTF-8 encoding (handles special characters like quotes or accents).

csv.DictWriter(f, fieldnames=["text", "author"]) creates a writer object that can write dictionaries with keys "text" and "author" as columns.

w.writeheader() writes the header row to the CSV file.

w.writerows(all_quotes) writes all the dictionaries in all_quotes as rows in the CSV.

open("quotes.json", "w", encoding="utf-8") opens a file named quotes.json for writing.

json.dump(all_quotes, f, ensure_ascii=False, indent=2) writes the list of dictionaries to the JSON file.

ensure_ascii=False preserves non-ASCII characters correctly.

indent=2 makes the JSON formatted and easy to read.

```python
import csv, json  # Import CSV and JSON libraries for saving data in these formats
# CSV
with open("quotes.csv", "w", newline="", encoding="utf-8") as f:
    w = csv.DictWriter(f, fieldnames=["text", "author"])
    w.writeheader()
    w.writerows(all_quotes)

# JSON
with open("quotes.json", "w", encoding="utf-8") as f:
    json.dump(all_quotes, f, ensure_ascii=False, indent=2)
```
