```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
# Step 1: Fetch the webpage
url = "https://books.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
```

```python

# Step 2: Extract all book containers
books = soup.select("article.product_pod")
```

```python
# Step 3: Extract title & price with regex cleanup
titles = []
prices = []

for book in books:
    # Title
    title = book.h3.a["title"]
    
    # Raw price (e.g. "£51.77")
    raw_price = book.find("p", class_="price_color").text
    
    # Regex to extract numeric value
    match = re.search(r"[\d]+\.\d+", raw_price)
    price = float(match.group()) if match else None
    
    titles.append(title)
    prices.append(price)
```

```python
# Step 4: Store in DataFrame
df = pd.DataFrame({
    "Title": titles,
    "Price": prices
})

```

```python
# Step 5: Preview & Save
print(df.head())  # display first 5 rows
df.to_csv("books_page1.csv", index=False)
```
