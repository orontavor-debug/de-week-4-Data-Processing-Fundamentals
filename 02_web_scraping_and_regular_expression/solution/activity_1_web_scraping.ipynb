```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Step 1: Fetch the IMDB Top 250 page
url = "https://www.imdb.com/chart/top/"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
}
response = requests.get(url, headers=headers)
```

```python

# Step 2: Parse the HTML
soup = BeautifulSoup(response.text, "html.parser")
```

```python
print(soup)
```

```python
# Step 3: Extract all movie rows
rows = soup.select("li.ipc-metadata-list-summary-item") 
```

```python
print(rows)
```

```python
# Step 4: Extract Title and Rating for the first 50 movies
titles = []
ratings = []

for row in rows[:25]:
    # Title
    title_tag = row.select_one("h3")
    title = title_tag.get_text(strip=True) if title_tag else None
    
    # Rating
    rating_tag = row.select_one("span.ipc-rating-star--rating")
    rating = rating_tag.get_text(strip=True) if rating_tag else None
    
    titles.append(title)
    ratings.append(float(rating) if rating else None)
```

```python
# Step 5: Store in Pandas DataFrame
df = pd.DataFrame({
    "Title": titles,
    "Rating": ratings
})
```

```python
print(df.count())
```

```python
# Step 6: Preview top 10 movies
print(df.head(10))
```

```python
# Step 7: Save to CSV
df.to_csv("top25_movies.csv", index=False)
print("\n Saved as top25_movies.csv")
```
