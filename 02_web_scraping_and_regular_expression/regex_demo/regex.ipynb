```python
# regex_demo.py
"""
This script demonstrates common regex syntaxes in Python.
Each pattern is shown with:
- Explanation of what it does
- Example text
- Matches produced
"""

import re

examples = [
    {"pattern": r"c.t", "text": "cat cut cot cbt ctt ct", "desc": "Matches any 'c' + any char + 't'"},
    {"pattern": r"a*b", "text": "b ab aab aaab", "desc": "Zero or more 'a' before b"},
    {"pattern": r"a+b", "text": "ab aab aaab b", "desc": "One or more 'a' before b"},
    {"pattern": r"a?b", "text": "b ab aab", "desc": "Zero or one 'a' before b"},
    {"pattern": r"[A-Za-z]", "text": "A 4 z _", "desc": "Matches any letter"},
    {"pattern": r"[^abc]", "text": "abcdxyz", "desc": "Matches any char except a,b,c"}, # ^ with a [] -> excep this character mentioned in side [] it will match with all
    {"pattern": r"^Hello", "text": "Hello world\nSay Hello", "desc": "Matches start of string"}, # start of the string
    {"pattern": r"world$", "text": "Hello world", "desc": "Matches end of string"},
    {"pattern": r"\d+", "text": "Order 12 and 345", "desc": "One or more digits"},
    {"pattern": r"\w+", "text": "hello_user 123", "desc": "Word characters"},
    {"pattern": r"(\d{2})/(\d{2})/(\d{4})", "text": "Date: 12/05/2024", "desc": "Captures DD/MM/YYYY"},
]

print("REGEX DEMONSTRATION:\n")
# \w -> alphanumeric along with underscore
for ex in examples:
    print(f"Pattern: {ex['pattern']}")
    print(f"Description: {ex['desc']}")
    print(f"Text: '{ex['text']}'")
    matches = re.findall(ex['pattern'], ex['text']) # find all the occurences of pattern within text and give matching words/text
    print(f"Matches: {matches}\n")

print("Done.")
```

re.match(pattern, string[, flags])

What it does (short): Attempts to match pattern only at the beginning of string.
Equivalent to implicitly anchoring the pattern with ^ at position 0.

Return value: a Match object if match succeeds at the start, otherwise None.

Step-by-step behavior

Python tries to match the regex at index 0 of the string.

If the pattern can match a prefix of the string → returns Match.

If not → returns None

```python
re.match(r"\d+", "123abc")    # Match object (matches "123")
```

```python
re.match(r"\d+", "abc123")    # None (digits are not at start)
```

```python
re.match(r"", "abc")          # Match object (empty string matched at pos 0)
# edge case empty string at position 0
```

Edge cases & notes

An empty pattern ("") always matches at position 0, returning a Match for the empty string.

re.match(r"^abc", "abc\nabc") matches the first abc only; re.match will not find later abc because it only looks at start.

To match at the start of each line use re.search with re.M + ^ (or iterate lines) — re.match does not search beyond position 0 even with re.M.

If the pattern is invalid, re.error is raised.

Passing a non-string pattern or string will typically raise TypeError.

re.search(pattern, string[, flags])

What it does (short): Scans the string and returns the first occurrence of the pattern anywhere in the string.

Return value: a Match object for the first match, or None if not found.

Step-by-step behavior

Engine scans left-to-right until it finds a substring that fits the pattern.

Returns the first Match object encountered.

If none found → None.

```python
re.search(r"\d+", "abc123xyz123131")  # Match for "123"
```

```python
re.search(r"cat", "doggo")      # None
```

Edge cases & notes

re.search(r"^foo", ...) still checks the beginning of the string unless re.M is used.

If pattern can match empty string, re.search may return a zero-length match (position where it first can match).

Use .start() / .end() on the Match to get indices.

If pattern is invalid → re.error.

re.findall(pattern, string[, flags])

What it does (short): Returns all non-overlapping matches as a list of strings (or tuples if there are capture groups).

Return value:

If the pattern has no capturing groups → list[str] (each matched substring).

If the pattern has one capturing group → list[str] (the group contents).

If the pattern has multiple groups → list[tuple] where each tuple contains the groups for that match.

If there are no matches → [] (empty list).

```python
re.findall(r"\d+", "abc123xyz456")         # ["123", "456"]
```

```python
re.findall(r"(\d+)-([A-Za-z]+)", "12-ab 34-cd")
# -> [("12","ab"), ("34","cd")]
```

```python
re.findall(r"cat", "no cats here")         # []
```

```python
re.findall('aba', 'ababa')
```

Edge cases & notes

Non-overlapping: re.findall('aba', 'ababa') returns ['aba'] — it will not return overlapping second 'aba'. To get overlapping matches use lookahead: (?=(aba)).

Empty matches: If a pattern can match an empty string, findall will include those empty matches, which can be surprising. Example: re.findall(r".*?", "abc") may include many empty strings. Be cautious.

Groups vs no-groups: If you only want full matches even with groups, use non-capturing groups (?:...), or use re.finditer and call .group(0).

Performance: findall returns results in memory; for large text use finditer.

re.finditer(pattern, string[, flags])

What it does (short): Returns an iterator of Match objects for each non-overlapping match.

Return value: an iterator; each iteration yields a Match object.

Examples

```python
[m.group() for m in re.finditer(r"\d+", "a1 b22 c333")]
# -> ["1", "22", "333"]
```

```python
for m in re.finditer(r"(\d+)-([A-Za-z]+)", "12-ab 34-cd"):
    print(m.group(0), m.group(1), m.group(2), m.span())

```

Edge cases & notes

Use this when you need match positions (.start(), .end()), groups, or want to process matches one by one (memory efficient).

Still non-overlapping by default.

Zero-length match caution: if the pattern matches empty strings repeatedly, the iterator will still progress — the engine takes special care to advance the position to avoid infinite loops, but patterns that match empty strings can be tricky to reason about.

finditer is preferred when you need indices, groups, or to avoid building a large list.

re.sub(pattern, repl, string[, count=0, flags=0])

What it does (short): Replaces all matches of pattern with repl and returns the new string. Use count to limit replacements.

Return value: the new string with substitutions applied.

repl can be:

a string, possibly containing backreferences like \1 or \g<name>

a function that receives a Match and returns a replacement string

```python
re.sub(r"\d+", "#", "ID123Code456")       # "ID#Code#"

# re.sub(pattern, replacementstring, mainstring)
```

```python
re.sub(r"(\w+) (\w+)", r"\2, \1", "John Doe")  # "Doe, John"
```

```python
# using function:
def repl(m): return f"<num:{m.group()}>"
re.sub(r"\d+", repl, "a1b22")    
```

Edge cases & notes

count=1 replaces only first occurrence.

Backslash handling: in replacement strings, use raw strings or double escapes. Example: r"\1" for group 1.

If pattern not found → original string is returned unchanged.

If repl is a function, it’s called with each Match; useful for formatting with groups or complex logic.

Invalid pattern → re.error.

Watch for unintended backreference expansion; prefer \g<name> for clarity with numbered groups >9.

re.split(pattern, string[, maxsplit=0, flags=0])

What it does (short): Splits string by occurrences of pattern and returns a list of substrings.

Return value: list of strings. If the pattern contains capturing groups, the captures are included in the result.

Examples

```python
re.split(r"\s+", "one two   three")  # ["one", "two", "three"]
```

```python
re.split(r"(\d+)", "a1bad22c")         # ["a", "1", "b", "22", "c"]  (captured groups are included)
```

```python
re.split(r",", "a,b,c", maxsplit=1)  # ["a", "b,c"]
```

Edge cases & notes

Leading/trailing separators produce empty strings: re.split(r",", ",a,b,") → ['', 'a', 'b', ''].

If no match → returns a 1-element list containing the original string.

Capturing groups in the splitting pattern are returned in the list — sometimes handy, sometimes surprising.

If pattern can match empty strings it can produce many empty pieces — be careful.

re.compile(pattern[, flags])

What it does (short): Compiles a regex pattern into a Pattern object so you can reuse it efficiently.

Return value: a Pattern object with methods match, search, findall, finditer, sub, split, etc.

Examples

```python
pat = re.compile(r"\d+")

pat.findall("a1b22c")       # ["1","22"]
```

```python
pat.search("x12y").group()  # "12"
```

Edge cases & notes

Use re.compile when the same pattern is used repeatedly — it slightly improves performance.

Flags like re.I, re.M, re.S can be provided: re.compile(r"^hello", re.I|re.M).

Invalid pattern → re.error.

A compiled pattern can be pickled? Not reliably across Python versions — generally don't rely on pickling compiled patterns.

1. Price

Matches the literal word:

Price


Regex treats plain text as exact characters unless escaped or special.

2. \s*

\s = any whitespace character (space, tab, newline)
* = zero or more occurrences

So \s* means:

0 or more spaces/tabs


Examples it matches:

Price:

Price :

Price :

3. :

Matches the literal colon character.

4. \s* (again)

Allows spaces AFTER the colon.

Matches:

:19.99

: 19.99

: 19.99

5. (\d+\.\d{2}) — CAPTURE GROUP

This is the important part, capturing the numeric price.

Breakdown:

( ) = capturing parentheses → saved as group(1)

\d+ = one or more digits
Examples: 1, 19, 123

\. = literal dot (decimal point)
(Must use \. because . normally means “any character”)

\d{2} = exactly two digits
Ensures two decimal places (.99)

So this captures prices like:

19.99
9.50
120.75


But not:

19.9 (only 1 decimal)

19.999 (3 decimals)

$19.99 (symbol is outside)

6. \$

Matches the literal dollar sign.

Why backslash needed?

$ in regex means "end of string"

So we escape it: \$

Putting it all together

The pattern:

Price\s*:\s*(\d+\.\d{2})\$

```python
import re
html_content = "<p>Price: 19.99$</p>"
pattern = r"Price\s*:\s*(\d+\.\d{2})\$"
match = re.search(pattern, html_content)
if match:
    print("Price found:", match.group(1))  # Output: 19.99

# explain this expression of pattern
```

1. http

Matches the literal text:

http

**2. [s]?

[s]? means:

Match optional 's'

[] is a character set (here containing only s)

? means 0 or 1 occurrence

So it matches:

"http"

"https"

This allows both HTTP and HTTPS URLs.

3. ://

Matches the literal sequence:

://


This is the standard URL protocol separator.

4. [^\s\"'>]+ — the URL body

This part captures the rest of the URL.

Let’s break it down:

A. [] → character class

Inside brackets we define allowed or prohibited characters.

B. ^ inside the class = NOT

^ at the start of a character class means negate / exclude.

C. Characters excluded:

\s → whitespace (spaces, tabs, newlines)

\" → double quote

' → single quote

> → angle bracket

< → angle bracket (implicitly not included, but usually paired)

So the pattern is saying:

Match one or more + characters that are NOT whitespace or " ' >
(These characters usually terminate HTML attributes.)

D. +

Matches one or more allowed characters to form the uninterrupted URL.

Final Meaning
✔️ The regex matches:

http://example.com

https://example.com/page

https://example.com/page?query=1&id=2

anything starting with http or https up to the first space or quote

```python
import re

html = '<a href="https://example.com/page1">Link 1</a>'
url_pattern = r"http[s]?://[^\s\"'>]+"
urls = re.findall(url_pattern, html)
print(urls)
```
